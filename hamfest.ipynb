```python
hams = pd.read_html('http://www.hamfestindia2018.com/Delegates/List?start=0&end=1400')[0]
```

```python
hams
```

```python
hams = hams.rename(columns=hams.iloc[0])
hams = hams.drop([0])
```

```python
hams['Gender']
```

```python
gend = hams['Gender'].value_counts()
fig1,ax1 = plt.subplots()
plt.pie(gend, explode = (0,0.2), labels = ('Old Man', 'Young Lady'),autopct='%1.2f%%',
       shadow = True, startangle = 50)
```
